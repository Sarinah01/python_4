```python
import networkx as nx
```

```python
g =nx.Graph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
nx.draw(g,with_labels=True,node_size=5000, node_color='red')
```

```python
g =nx.DiGraph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,1)
nx.draw(g,with_labels=True,node_size=5000, node_color='red')
```

```python
g =nx.DiGraph()
g.add_nodes_from([1,2,3,4,5])
g.add_edges_from([(1,2),(2,3),(3,4),(1,1),(1,4)])
nx.draw(g,with_labels=True,node_size=1000, node_color='red')
```

```python
import pandas as pd
import plotly.express as px

df_orders = pd.DataFrame({
    "City": ["Ahmedabad", "Ahmedabad", "Ahmedabad",
             "Mumbai", "Mumbai", "Mumbai",
             "Delhi", "Delhi", "Delhi"],
    "Category": ["Electronics", "Clothing", "Grocery",
                 "Electronics", "Clothing", "Grocery",
                 "Electronics", "Clothing", "Grocery"],
    "Orders": [120, 90, 150, 200, 140, 180, 170, 110, 160]
})

print(df_orders)
px.bar(data_frame =df_orders,x="City",y="Orders",color="Category",barmode="group",pattern_shape="Category",hover_data="Category",text_auto=True)
```
