```python
url = "https://subslikescript.com/movies_letter-V"
response = requests.get(url)
response.status_code
sp = BeautifulSoup(response.text,'html.parser')
box = sp.find('article', class_='main-article')
anchor =box.find_all('a')
```

```python
links = []
names = []
for i in anchor:
    names.append(i.text.strip())
    links.append("https://subslikescript.com"+ i['href'])
links
names
```

```python
for x ,y in zip(names,links):
    url = y
    response = requests.get(url)
    sp = BeautifulSoup(response.text,'html.parser')
    script = sp.find('div', class_="full-script").text
    with open(f"{x}.html","w") as f:
        f.write(script)
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
response = requests.get(url)
response.status_code
sp = BeautifulSoup(response.text,'html.parser')
sp.prettify
rows = sp.find_all('div',class_="product-wrapper card-body")
len(rows)
rows
```

```python
d = {"name":[], "price":[],"reviews":[]}
for i in rows:
    d['name'].append(i.find('a').text.strip())
    d['price'].append(i.find('span').text.strip())
    d['reviews'].append(i.find('p', class_='review-count float-end').text.strip())
df = pd.DataFrame(d)
df
```

```python
df.to_csv('data.csv', index=False)
```
